# **Introduction to Matplotlib**

```python
# Import Library
import matplotlib.pyplot as plt
```

```python
X = np.arange(1,20)
Y = X**2
```

```python
print(X)
print(Y)
```

```python
# .plot(x, y), color=

plt.plot(X, Y, color='red')
```

```python
# .xlabel(), .ylabel(), .title(), '.'

plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('X^2')
plt.title('My Plot')
```

```python
# .bar(), .barh(), .polar(), .step()

plt.scatter(X, Y)
```

```python
# .figure(figsize=()), .subplot(2,2,n)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('X^2')
plt.title('My Plot')

plt.subplot(2,2,2)
plt.plot(X, np.sin(X))
plt.xlabel('X')
plt.ylabel('sin X')
plt.title('My Plot')

plt.subplot(2,2,3)
plt.plot(X, np.sqrt(X))
plt.xlabel('X')
plt.ylabel('sqrt X')
plt.title('My Plot')

plt.subplot(2,2,4)
plt.plot(X, np.log10(X))
plt.xlabel('X')
plt.ylabel('Log X')
plt.title('My Plot')
```

```python
# Figure, Axes .add_axes, set_xlabels , ....
X = np.linspace(0,2,50)
Y = np.sin(X)
Figure = plt.figure()
Axes = Figure.add_axes([0, 0, 2, 1])
Axes.plot(X,Y)
Axes.set_xlabel('X')
Axes.set_ylabel('Sin X')
```

```python
# .subplots()

Figure, Axes = plt.subplots(nrows=2, ncols=1)
Axes[0].plot(X, X**2)
Axes[1].plot(X,np.log10(X))
```
